 EXPLORATORY DATA ANALYSIS

#GENDER

```python
ax=sns.countplot(x="Gender" , data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_gen=data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)

ay=sns.barplot(x='Gender', y='Amount' , data=sales_gen)
for bars in ay.containers:
    ay.bar_label(bars)
plt.show()
```

*** from above graph we can see that most of the buyers are females and evn the have purchasing power of females is greater than men

#AGE 

```python
ax=sns.countplot(x='Age Group', hue='Gender' ,data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

***from above graph we can see that most of buyers are of age group between 26-35

#STATE

```python
sales_state=data.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
az=sns.barplot(data=sales_state, x='State' , y='Orders')
for bars in az.containers:
    az.bar_label(bars)
plt.show()
```

```python
sales_state=data.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
az=sns.barplot(data=sales_state, x='State' , y='Amount')
for bars in az.containers:
    az.bar_label(bars)
plt.show()
```

#MARITAL STATUS

```python
ax=sns.countplot(data=data, x="Marital_Status")
sns.set(rc={'figure.figsize':(5,2)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=data, x= "Marital_Status" , y="Amount", hue='Gender' )
```

#OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=data, x="Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=data.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
az=sns.barplot(data=sales_state, x='Occupation' , y='Amount')

plt.show()
```

#product category

```python
sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(data=data, x="Product_Category")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,7)})
az=sns.barplot(data=sales_state, x='Product_Category' , y='Amount')

plt.show()
```

from above graph we can see that most of the sold products are food > clothing > elctronics category

```python
sales_state=data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
az=sns.barplot(data=sales_state, x='Product_ID' , y='Orders')

plt.show()
```
